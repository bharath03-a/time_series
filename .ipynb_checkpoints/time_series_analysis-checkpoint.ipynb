```python
# Getting yearly sum, quarterly sum
oil_yearly_sum = oil.resample('A').mean()
oil_yearly_sum.plot()
plt.title("Oil Prices yearly average")
plt.xlabel("year")
plt.ylabel("Oil Price")
```

```python
# Getting yearly sum, quarterly sum
oil_qrly_sum = oil.resample('Q').mean()
oil_qrly_sum.plot()
plt.title("Oil Prices quarterly  average")
plt.xlabel("year")
plt.ylabel("Oil Price")
```

```python
train_oil = oil[oil.index.year <= 2016]
test_oil = oil[oil.index.year > 2016]
```

```python
print(train_oil.shape)
print(test_oil.shape)
```

```python
train_oil['dcoilwtico'].plot(figsize = (12, 5))
test_oil['dcoilwtico'].plot(figsize = (12, 5))
plt.legend(['Training', 'Testing'])
plt.show()
```

```python
train_time = [i+1 for i in range(len(train_oil))]
test_time = [i+len(train_oil)+1 for i in range(len(test_oil))]
```

```python
LR_train = train_oil.copy()
LR_test = test_oil.copy()
```

```python
LR_train['time'] = train_time
LR_test['time'] = test_time
```

```python
lr = LinearRegression()
lr.fit(LR_train[['time']], LR_train['dcoilwtico'].values)
```

```python
test_pred_model = lr.predict(LR_test[['time']])
LR_test['forecast'] = test_pred_model
```

```python
plt.figure(figsize = (12, 6))
plt.plot(train_oil['dcoilwtico'], label = 'train')
plt.plot(test_oil['dcoilwtico'], label = 'test')
plt.plot(LR_test['forecast'], label = 'reg on time_test data')
plt.legend(loc = 'best')
```

```python
def mape(act, pred):
    return round((np.mean(abs(act - pred)/act))*100, 2)
```

```python
mape_model_test = mape(test_oil['dcoilwtico'].values, test_pred_model)
print(f"MAPE is {mape_model_test}%")
```
